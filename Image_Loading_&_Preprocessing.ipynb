```python
import os
from PIL import Image
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_dir='/content/drive/MyDrive/Medical_image_project/Alzheimer_s Dataset'
```

```python
files = os.listdir(image_dir)
print("All files in the directory:", files)
```

Recursive Search in Subdirectories

```python
image_files = []
for root, dirs, files in os.walk(image_dir):
    for file in files:
        if file.lower().endswith('.jpg'):
            image_files.append(os.path.join(root, file))

print("Found JPG files:", image_files)
```

```python
for i, file_path in enumerate(image_files[:5]):
    img = Image.open(file_path)
    plt.figure(figsize=(5,5))
    plt.imshow(img, cmap='gray')
    plt.title(f"Image {i+1}: {os.path.basename(file_path)}")
    plt.axis('off')
    plt.show()
```

```python
for file_path in image_files[:5]:
    img = Image.open(file_path)
    print(f"\nFile: {file_path}")
    print("Format:", img.format)
    print("Size:", img.size)
    print("Mode:", img.mode)
    print("Image Info:", img.info)
```

```python
import numpy as np

# Normalize images
normalized_images = []
for file_path in image_files[:5]:
    img=Image.open(file_path)
    img_array = np.array(img) / 255.0  # Normalize pixel values to [0, 1]
    normalized_images.append(img_array)
```
